<h5>2) Requesting Data </h5>

```python
from bs4 import BeautifulSoup
import requests
source=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,"lxml")
```

<h5>3)Converting Data into Dataframe using Pandas  <h5>

```python
table=soup.find('table')
table_rows = table.find_all('tr')
res=[]
column_names=["PostalCode", "Borough", "Neighborhood"]
df = pd.DataFrame(columns=column_names)
for tr in table_rows:
    td = tr.find_all('td')
    i=0
    postalcode=""
    borough=""
    neighborhood=""
    neighborhood_list=""
    for d in td:
        if(i==0):
            postalcode=d.text.strip()
            i=i+1
            continue
        if(i==1):
            borough=d.text.strip()
            i=i+1
            if(borough=="Not assigned" or borough==""):
                i=-1
                break
        if(i==2):
            neighborhood=d.text.strip()
            neighborhood_list=neighborhood.split(' / ')
            neighborhood=','.join(neighborhood_list)
    if(i!=-1):
        df = df.append({'PostalCode':postalcode,'Borough': borough,'Neighborhood': neighborhood}, ignore_index=True)
```

<h5>4)Removing empty first row and Displaying the result<h5>

```python
df.drop(df.index[0], inplace=True)
```
